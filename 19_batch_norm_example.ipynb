```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, BatchNormalization, Dropout, ReLU
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler
from tensorflow.keras.optimizers import Adam
import numpy as np
```

```python
# Build the model
model = Sequential()

# First Dense Layer
model.add(Dense(2, input_dim=2))
model.add(BatchNormalization())  # After Dense Layer
model.add(ReLU())  # Activation after Batch Norm
model.add(Dropout(0.1))  # Dropout layer to prevent overfitting

# Second Dense Layer
model.add(Dense(16))
model.add(BatchNormalization())  # After Dense Layer
model.add(ReLU())  # Activation after Batch Norm
model.add(Dropout(0.1))  # Dropout layer to prevent overfitting

# Second Dense Layer
model.add(Dense(16))
model.add(BatchNormalization())  # After Dense Layer
model.add(ReLU())  # Activation after Batch Norm
model.add(Dropout(0.1))  # Dropout layer to prevent overfitting

# Output Layer
model.add(Dense(1, activation='sigmoid'))

model.summary()
```

```python
# Compile the model with Adam optimizer
model.compile(loss='binary_crossentropy', optimizer=Adam(), metrics=['accuracy'])
```

```python

# EarlyStopping callback
# early_stopping = EarlyStopping(
#     monitor='val_loss',  # Watch validation loss
#     patience=10,  # Stop after 10 epochs of no improvement
#     restore_best_weights=True,  # Restore the weights of the best epoch
#     verbose=1
# )

# Learning Rate Scheduler callback
def lr_schedule(epoch):
    initial_lr = 0.001  # Starting learning rate
    drop_rate = 0.5  # Factor to reduce learning rate
    drop_every = 10  # Every 10 epochs, reduce the learning rate by a factor
    if epoch % drop_every == 0 and epoch > 0:
        return initial_lr * drop_rate
    return initial_lr

learning_rate_scheduler = LearningRateScheduler(lr_schedule, verbose=1)
```

```python
# Train the model with EarlyStopping and Learning Rate Scheduler
history3 = model.fit(
    X, y,  # Your input and labels
    epochs=200,  # Maximum number of epochs
    validation_split=0.2,  # Hold out 20% of data for validation
    callbacks=[learning_rate_scheduler]  # Apply the callbacks
)
```

```python
# Plot training and validation accuracy
plt.plot(history3.history['val_accuracy'], label='Validation Accuracy',color='black')
plt.plot(history3.history['accuracy'], label='Training Accuracy', color='green')

# Add labels and title
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training vs Validation Accuracy')
plt.legend()

# Show the plot
plt.show()
```

```python
# Plot training and validation accuracy
plt.plot(history3.history['loss'], label='Training Loss', color='green')
plt.plot(history3.history['val_loss'], label='Validation Loss',color='black')

# Add labels and title
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training vs Validation Accuracy')
plt.legend()

# Show the plot
plt.show()
```

```python
# Plot training and validation loss alongside accuracy
plt.figure(figsize=(11, 4))

# Accuracy plot
plt.subplot(1, 2, 1)
plt.plot(history3.history['val_accuracy'], label='Validation Accuracy',color='black')
plt.plot(history3.history['accuracy'], label='Training Accuracy', color='green')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.title('Training vs Validation Accuracy')
plt.legend()

# Loss plot
plt.subplot(1, 2, 2)
plt.plot(history3.history['loss'], label='Training Loss', color='green')
plt.plot(history3.history['val_loss'], label='Validation Loss',color='black')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.title('Training vs Validation Loss')
plt.legend()

# Show the plot
plt.tight_layout()
plt.show()
```
